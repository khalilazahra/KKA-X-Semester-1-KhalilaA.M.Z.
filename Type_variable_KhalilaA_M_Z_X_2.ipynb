#Assigment

```python
#single assigment
bb_fazel = 55
bb_dani = 60
print(bb_dani)
print(bb_fazel)
```

```python
#multiple assigment
bb_fazel, bb_dani = 55,60
print(bb_dani)
print(bb_fazel)
```

```python
nama_siswa_1, nama_siswa_2 ="deni","doni"
print(nama_siswa_1)
print(nama_siswa_2)
```

```python
#chain assigment
bb_hasbi = bb_dani = 60
print(bb_hasbi)
print(bb_dani)
```
